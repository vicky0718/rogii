```python
# Regex to extract leading number
import re

pattern = r"^(.*?)_(?:typewell|horizontal_wells)\.csv$"

def extract_well_hashes(directory):
    hashes = set()

    for filename in os.listdir(directory):

        match = re.match(pattern, filename)

        if match:
            hashes.add(match.group(1))

    return hashes

# Extract hashes
train_hashes = extract_well_hashes(TRAIN_DIR)
test_hashes = extract_well_hashes(TEST_DIR)

print(f"Train wells: {len(train_hashes)}")
print(f"Test wells: {len(test_hashes)}")

# Overlap
overlap = train_hashes.intersection(test_hashes)

print(f"Overlapping wells: {len(overlap)}")

if overlap:
    print("Sample overlaps:", list(overlap)[:10])
else:
    print("No overlap found.")
```

# Check eval zone length distribution
# How many MD rows are hidden per well? Min / median / max. Short vs long zones behave differently.

```python
pattern = r"^(.*?)_(?:typewell|horizontal_wells)\.csv$"

def get_eval_zone_lengths(directory):

    zone_lengths = []

    for filename in os.listdir(directory):

        if not filename.endswith(".csv"):
            continue

        match = re.match(pattern, filename)

        if not match:
            continue

        filepath = os.path.join(directory, filename)

        try:
            df = pd.read_csv(filepath)

            # Find hidden / eval rows
            # Usually target column contains NaNs in eval region
            null_counts = df.isnull().sum()

            # Column with most nulls assumed to be target
            target_col = null_counts.idxmax()

            hidden_mask = df[target_col].isnull()

            hidden_len = hidden_mask.sum()

            zone_lengths.append({
                "file": filename,
                "well_id": match.group(1),
                "hidden_rows": hidden_len
            })

        except Exception as e:
            print(f"Error reading {filename}: {e}")

    return pd.DataFrame(zone_lengths)

# Process train and test
train_eval = get_eval_zone_lengths(TRAIN_DIR)
test_eval = get_eval_zone_lengths(TEST_DIR)

# Combine
all_eval = pd.concat([
    train_eval.assign(dataset="train"),
    test_eval.assign(dataset="test")
])

# Summary statistics
summary = all_eval.groupby("dataset")["hidden_rows"].agg([
    "count",
    "min",
    "median",
    "max",
    "mean"
])

print(summary)

# Optional: inspect longest / shortest zones
print("\nTop longest hidden zones:")
print(all_eval.sort_values("hidden_rows", ascending=False).head(10))

print("\nTop shortest hidden zones:")
print(all_eval.sort_values("hidden_rows").head(10))
```

```python


TRAIN_DIR = "../data/train"

files = glob.glob(os.path.join(TRAIN_DIR, "*horizontal_well.csv"))

print(f"Found {len(files)} horizontal well files")

if len(files) == 0:
    raise ValueError(
        f"No files found in {TRAIN_DIR}"
    )

results = []

for fp in tqdm(files):

    try:
        well_name = os.path.basename(fp)

        df = pd.read_csv(fp)

        required_cols = ["TVT", "TVT_input"]

        missing_cols = [
            c for c in required_cols
            if c not in df.columns
        ]

        if len(missing_cols) > 0:
            print(f"Skipping {well_name} - missing columns {missing_cols}")
            continue

        known_mask = df["TVT_input"].notna()

        known = df.loc[
            known_mask,
            ["TVT", "TVT_input"]
        ].copy()

        if len(known) == 0:
            print(f"Skipping {well_name} - no known zone")
            continue

        diff = known["TVT"] - known["TVT_input"]

        abs_diff = diff.abs()

        exact_match = (abs_diff == 0).all()

        allclose_1e6 = np.allclose(
            known["TVT"],
            known["TVT_input"],
            atol=1e-6,
            rtol=0
        )

        allclose_1e4 = np.allclose(
            known["TVT"],
            known["TVT_input"],
            atol=1e-4,
            rtol=0
        )

        max_abs_diff = abs_diff.max()
        mean_abs_diff = abs_diff.mean()

        last_known_idx = known.index[-1]

        ps_diff = abs(
            df.loc[last_known_idx, "TVT"] -
            df.loc[last_known_idx, "TVT_input"]
        )

        results.append({
            "well": well_name,
            "known_rows": len(known),

            "exact_match": exact_match,
            "allclose_1e6": allclose_1e6,
            "allclose_1e4": allclose_1e4,

            "max_abs_diff": max_abs_diff,
            "mean_abs_diff": mean_abs_diff,

            "ps_diff": ps_diff
        })

    except Exception as e:
        print(f"Error processing {fp}")
        print(e)

summary = pd.DataFrame(results)

print("\n")
print("=" * 80)
print("SUMMARY")
print("=" * 80)

print(f"Processed wells: {len(summary)}")

if len(summary) > 0:

    print(f"Exact copy wells: {summary['exact_match'].sum()}")
    print(f"1e-6 tolerant wells: {summary['allclose_1e6'].sum()}")
    print(f"1e-4 tolerant wells: {summary['allclose_1e4'].sum()}")

    print("\nMax absolute difference:")
    print(summary["max_abs_diff"].describe())

    print("\nPS difference:")
    print(summary["ps_diff"].describe())

else:
    print("No wells processed")
```

# Check typewell TVT range vs horizontal TVT range
# If horizontal TVT goes outside typewell range, fallback interpolation is extrapolation (dangerous).

```python
horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

results = []

for h_path in tqdm(horizontal_files):

    h_name = os.path.basename(h_path)

    well_id = h_name.replace(
        "__horizontal_well.csv",
        ""
    )

    tw_matches = glob.glob(
        os.path.join(
            TRAIN_DIR,
            f"{well_id}__typewell*.csv"
        )
    )

    if len(tw_matches) != 1:
        continue

    tw_path = tw_matches[0]

    h = pd.read_csv(h_path)
    tw = pd.read_csv(tw_path)

    pred_mask = h["TVT_input"].isna()

    # Safety check
    if pred_mask.sum() == 0:
        continue

    pred_tvt_min = h.loc[pred_mask, "TVT"].min()
    pred_tvt_max = h.loc[pred_mask, "TVT"].max()

    tw_tvt_min = tw["TVT"].min()
    tw_tvt_max = tw["TVT"].max()

    below_range_ft = max(
        0.0,
        tw_tvt_min - pred_tvt_min
    )

    above_range_ft = max(
        0.0,
        pred_tvt_max - tw_tvt_max
    )

    total_outside_ft = (
        below_range_ft +
        above_range_ft
    )

    requires_extrapolation = (
        total_outside_ft > 0
    )

    results.append({
        "well_id": well_id,

        "pred_tvt_min": pred_tvt_min,
        "pred_tvt_max": pred_tvt_max,

        "typewell_tvt_min": tw_tvt_min,
        "typewell_tvt_max": tw_tvt_max,

        "below_range_ft": below_range_ft,
        "above_range_ft": above_range_ft,

        "total_outside_ft": total_outside_ft,

        "requires_extrapolation": requires_extrapolation
    })

summary = pd.DataFrame(results)

print("=" * 80)
print("TYPEWELL TVT COVERAGE CHECK (PREDICTION ZONE)")
print("=" * 80)

print(f"Processed wells: {len(summary)}")

if len(summary) == 0:
    print("No wells processed.")
else:

    n_extrap = summary[
        "requires_extrapolation"
    ].sum()

    print(
        f"Wells requiring extrapolation: "
        f"{n_extrap}"
    )

    print(
        f"Wells fully covered: "
        f"{len(summary) - n_extrap}"
    )

    print(
        f"Percent requiring extrapolation: "
        f"{100 * n_extrap / len(summary):.2f}%"
    )

    offenders = summary[
        summary["requires_extrapolation"]
    ].copy()

    if len(offenders) > 0:

        print("\nOutside-range statistics:")
        print(
            offenders[
                "total_outside_ft"
            ].describe()
        )

        print("\nTop 20 worst offenders:")
        print(
            offenders.sort_values(
                "total_outside_ft",
                ascending=False
            ).head(20)
        )

summary.to_csv(
    "prediction_zone_typewell_tvt_coverage.csv",
    index=False
)
```

```python
results = []

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for h_path in tqdm(horizontal_files):

    h_name = os.path.basename(h_path)

    well_id = h_name.replace(
        "__horizontal_well.csv",
        ""
    )

    tw_matches = glob.glob(
        os.path.join(
            TRAIN_DIR,
            f"{well_id}__typewell*.csv"
        )
    )

    if len(tw_matches) != 1:
        continue

    tw_path = tw_matches[0]

    h = pd.read_csv(h_path)
    tw = pd.read_csv(tw_path)

    # --------------------------------------------------
    # Length metrics
    # --------------------------------------------------

    known_mask = h["TVT_input"].notna()

    known_len = known_mask.sum()
    total_len = len(h)
    pred_len = total_len - known_len

    known_ratio = known_len / total_len

    # --------------------------------------------------
    # GR correlation vs typewell
    # --------------------------------------------------

    corr_vs_typewell = np.nan

    try:

        known = h.loc[
            known_mask,
            ["TVT", "GR"]
        ].dropna()

        if len(known) > 20:

            interp_gr = interp1d(
                tw["TVT"],
                tw["GR"],
                kind="linear",
                bounds_error=False,
                fill_value=np.nan
            )

            tw_gr_at_h_tvt = interp_gr(
                known["TVT"].values
            )

            valid = np.isfinite(
                tw_gr_at_h_tvt
            )

            if valid.sum() > 20:

                corr_vs_typewell = np.corrcoef(
                    known["GR"].values[valid],
                    tw_gr_at_h_tvt[valid]
                )[0, 1]

    except Exception:
        pass

    results.append({
        "well_id": well_id,

        "known_len": known_len,
        "pred_len": pred_len,
        "total_len": total_len,

        "known_ratio": known_ratio,

        "gr_corr_vs_typewell":
            corr_vs_typewell
    })

summary = pd.DataFrame(results)

# =====================================================
# Percentiles
# =====================================================

p01 = summary["known_len"].quantile(0.01)
p05 = summary["known_len"].quantile(0.05)

print("=" * 80)
print("KNOWN-ZONE LENGTH DISTRIBUTION")
print("=" * 80)

print(
    summary["known_len"].describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\nP01 =", p01)
print("P05 =", p05)

# =====================================================
# P01 wells
# =====================================================

p01_wells = (
    summary
    .loc[
        summary["known_len"] <= p01
    ]
    .sort_values("known_len")
)

print("\n")
print("=" * 80)
print("P01 WELLS")
print("=" * 80)

print(
    p01_wells[
        [
            "well_id",
            "known_len",
            "pred_len",
            "known_ratio",
            "gr_corr_vs_typewell"
        ]
    ]
)

# =====================================================
# P05 wells
# =====================================================

p05_wells = (
    summary
    .loc[
        summary["known_len"] <= p05
    ]
    .sort_values("known_len")
)

print("\n")
print("=" * 80)
print("P05 WELLS")
print("=" * 80)

print(
    p05_wells[
        [
            "well_id",
            "known_len",
            "pred_len",
            "known_ratio",
            "gr_corr_vs_typewell"
        ]
    ]
)

# =====================================================
# Compare groups
# =====================================================

summary["length_group"] = "Normal"

summary.loc[
    summary["known_len"] <= p05,
    "length_group"
] = "Short_P05"

summary.loc[
    summary["known_len"] <= p01,
    "length_group"
] = "VeryShort_P01"

print("\n")
print("=" * 80)
print("GROUP COMPARISON")
print("=" * 80)

print(
    summary
    .groupby("length_group")
    .agg({
        "known_len": ["count", "mean", "median"],
        "pred_len": ["mean", "median"],
        "known_ratio": ["mean"],
        "gr_corr_vs_typewell": ["mean", "median"]
    })
)

summary.to_csv(
    "pre_ps_length_alignment_analysis.csv",
    index=False
)
```

```python

from scipy.interpolate import interp1d

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

results = []

for h_path in tqdm(horizontal_files):

    h_name = os.path.basename(h_path)

    well_id = h_name.replace(
        "__horizontal_well.csv",
        ""
    )

    tw_matches = glob.glob(
        os.path.join(
            TRAIN_DIR,
            f"{well_id}__typewell*.csv"
        )
    )

    if len(tw_matches) != 1:
        continue

    try:

        h = pd.read_csv(h_path)
        tw = pd.read_csv(tw_matches[0])

        known_mask = (
            h["TVT_input"].notna()
            & h["GR"].notna()
        )

        known = h.loc[
            known_mask,
            ["TVT", "GR"]
        ].copy()

        # Sort typewell by TVT
        tw = tw.sort_values("TVT")

        # Remove duplicate TVT values if any
        tw = tw.drop_duplicates(
            subset=["TVT"],
            keep="first"
        )

        n_known = len(known)

        tw_tvt_min = tw["TVT"].min()
        tw_tvt_max = tw["TVT"].max()

        if n_known < 20:

            results.append({
                "well_id": well_id,
                "known_points": n_known,
                "overlap_points": 0,
                "corr": np.nan,
                "mae": np.nan,
                "status": "too_few_known_points"
            })

            continue

        interp_fn = interp1d(
            tw["TVT"].values,
            tw["GR"].values,
            kind="linear",
            bounds_error=False,
            fill_value=np.nan
        )

        interp_gr = interp_fn(
            known["TVT"].values
        )

        valid = (
            np.isfinite(interp_gr)
            &
            np.isfinite(
                known["GR"].values
            )
        )

        overlap_points = valid.sum()

        if overlap_points < 20:

            results.append({
                "well_id": well_id,
                "known_points": n_known,
                "overlap_points": overlap_points,
                "corr": np.nan,
                "mae": np.nan,
                "status": "too_few_overlap_points"
            })

            continue

        corr = np.corrcoef(
            known["GR"].values[valid],
            interp_gr[valid]
        )[0, 1]

        mae = np.mean(
            np.abs(
                known["GR"].values[valid]
                -
                interp_gr[valid]
            )
        )

        results.append({
            "well_id": well_id,
            "known_points": n_known,
            "overlap_points": overlap_points,
            "corr": corr,
            "mae": mae,
            "status": "ok"
        })

    except Exception as e:

        results.append({
            "well_id": well_id,
            "known_points": np.nan,
            "overlap_points": np.nan,
            "corr": np.nan,
            "mae": np.nan,
            "status": f"error: {str(e)[:80]}"
        })

summary = pd.DataFrame(results)

print("=" * 80)
print("CORRELATION DIAGNOSTICS")
print("=" * 80)

print("\nStatus counts:")
print(summary["status"].value_counts())

ok = summary[
    summary["status"] == "ok"
]

print("\nSuccessful wells:")
print(len(ok))

if len(ok):

    print("\nCorrelation distribution:")
    print(
        ok["corr"].describe(
            percentiles=[
                0.01,
                0.05,
                0.25,
                0.5,
                0.75,
                0.95,
                0.99
            ]
        )
    )

    print("\nMAE distribution:")
    print(
        ok["mae"].describe(
            percentiles=[
                0.01,
                0.05,
                0.25,
                0.5,
                0.75,
                0.95,
                0.99
            ]
        )
    )

print("\nProblem wells:")
print(
    summary[
        summary["status"] != "ok"
    ].head(50)
)

summary.to_csv(
    "gr_typewell_correlation_diagnostics.csv",
    index=False
)
```

# Plot per-well TVT trend along MD Classify wells:
### flat (TVT ~constant), dipping up, dipping down, reversing. Each class needs different treatment.

```python
# Compute TVT Trend Metrics Per Well

# This extracts the key quantities needed for classification.

from scipy.stats import linregress

results = []

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for fp in tqdm(horizontal_files):

    well_id = (
        os.path.basename(fp)
        .replace("__horizontal_well.csv", "")
    )

    df = pd.read_csv(fp)

    md = df["MD"].values
    tvt = df["TVT"].values

    # Global trend
    slope, intercept, r_value, _, _ = linregress(md, tvt)

    # Total TVT change
    tvt_change = tvt[-1] - tvt[0]

    # First-half vs second-half slope
    mid = len(df) // 2

    slope1 = linregress(
        md[:mid],
        tvt[:mid]
    ).slope

    slope2 = linregress(
        md[mid:],
        tvt[mid:]
    ).slope

    results.append({
        "well_id": well_id,

        "global_slope": slope,
        "r2": r_value**2,

        "tvt_change": tvt_change,

        "slope_first_half": slope1,
        "slope_second_half": slope2
    })

trend_df = pd.DataFrame(results)

print(trend_df.describe())

trend_df.to_csv(
    "tvt_trend_metrics.csv",
    index=False
)
```

```python
# Cell 2 — Classify Wells

# Classification rules:

# Flat
# |TVT change| < 50 ft
# Dipping Down
# TVT change > 50
# Dipping Up
# TVT change < -50
# Reversing
# first-half slope
# and
# second-half slope

# have opposite signs

trend_df = pd.read_csv(
    "tvt_trend_metrics.csv"
)

def classify(row):

    if (
        np.sign(row["slope_first_half"])
        !=
        np.sign(row["slope_second_half"])
    ):
        return "reversing"

    if abs(row["tvt_change"]) < 50:
        return "flat"

    if row["tvt_change"] > 0:
        return "dipping_down"

    return "dipping_up"

trend_df["trend_class"] = (
    trend_df.apply(
        classify,
        axis=1
    )
)

print(
    trend_df["trend_class"]
    .value_counts()
)

trend_df.to_csv(
    "tvt_trend_classification.csv",
    index=False
)
```

```python
# Cell 3 — Visualize Example Wells From Each Class

# This is the most useful part.


trend_df = pd.read_csv(
    "tvt_trend_classification.csv"
)

for trend_class in [
    "flat",
    "dipping_down",
    "dipping_up",
    "reversing"
]:

    subset = trend_df[
        trend_df["trend_class"]
        == trend_class
    ]

    if len(subset) == 0:
        continue

    example = subset.iloc[0]["well_id"]

    fp = glob.glob(
        os.path.join(
            TRAIN_DIR,
            f"{example}__horizontal_well.csv"
        )
    )[0]

    df = pd.read_csv(fp)

    plt.figure(figsize=(10,5))

    plt.plot(
        df["MD"],
        df["TVT"]
    )

    plt.title(
        f"{trend_class}: {example}"
    )

    plt.xlabel("MD")
    plt.ylabel("TVT")

    plt.grid(True)

    plt.show()
```

# Compute delta-TVT per MD foot in known zone
### This is the geological dip rate. Distribution tells you how fast TVT changes on average.

```python
# This is a very good EDA step and directly relevant to the forecasting problem.

# The goal is to quantify:

# ΔTVT / ΔMD

# Since you've already verified:

# MD step = 1 ft
# No duplicate MD

# this simplifies to:

# dTVT = TVT.diff()

# and represents the geological dip rate per foot drilled.

all_dip_rates = []
well_stats = []

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for fp in tqdm(horizontal_files):

    well_id = (
        os.path.basename(fp)
        .replace("__horizontal_well.csv", "")
    )

    df = pd.read_csv(fp)

    known = df.loc[
        df["TVT_input"].notna(),
        ["MD", "TVT"]
    ].copy()

    if len(known) < 2:
        continue

    dmd = known["MD"].diff()
    dtvt = known["TVT"].diff()

    dip_rate = dtvt / dmd

    dip_rate = dip_rate.replace(
        [np.inf, -np.inf],
        np.nan
    ).dropna()

    if len(dip_rate) == 0:
        continue

    all_dip_rates.extend(
        dip_rate.values
    )

    well_stats.append({
        "well_id": well_id,

        "mean_dip_rate":
            dip_rate.mean(),

        "median_dip_rate":
            dip_rate.median(),

        "std_dip_rate":
            dip_rate.std(),

        "p95_abs_dip_rate":
            np.percentile(
                np.abs(dip_rate),
                95
            ),

        "max_abs_dip_rate":
            np.max(
                np.abs(dip_rate)
            )
    })

all_dip_rates = np.array(all_dip_rates)

well_stats = pd.DataFrame(well_stats)

print("=" * 80)
print("GLOBAL DIP RATE DISTRIBUTION")
print("=" * 80)

print(
    pd.Series(all_dip_rates)
    .describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\n")
print("=" * 80)
print("PER-WELL MEAN DIP RATE")
print("=" * 80)

print(
    well_stats["mean_dip_rate"]
    .describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

well_stats.to_csv(
    "known_zone_dip_rate_stats.csv",
    index=False
)
```

```python
# This result is actually more interesting than I expected, and it reveals a subtle issue.

# First Observation: The median dip rate is not near zero

# You got:

# Median = 0.32 ft TVT / ft MD
# Mean   = 0.42 ft TVT / ft MD

# That means during the known zone:

# for every 100 ft drilled
# TVT increases ~32–42 ft

# That's substantial.

# This is consistent with what we saw earlier:

# Most wells start with a build section.
# TVT rises rapidly before flattening.
# Known zones include that build section.
# Second Observation: The distribution is contaminated by outliers

# Look at:

# min = -151.38
# max = +13.97

# Those are geologically impossible as true dip rates.

# A change of:

# 151 ft TVT
# in 1 ft MD

# cannot be a real well trajectory.

# This strongly suggests:

# TVT jumps,
# boundary artifacts,
# isolated bad rows,
# or formation/top transition discontinuities.

# The fact that:

# P01 = -0.06
# P99 = +1.30

# but

# min = -151

# means the extreme values are very rare.

# What we should really measure

# The current metric is mixing:

# Build section
# +
# Landing section
# +
# Lateral section

# The competition prediction happens after PS, and we've already learned:

# Second-half TVT is almost flat

# So the more useful EDA is:

# Dip rate in the last known zone

# For example:

# last_500ft_before_PS

# or

# last_1000ft_before_PS

# because that is the geological state immediately before forecasting.

# Recommended EDA
window = 500  # or 1000

known = df[df["TVT_input"].notna()].copy()

tail = known.tail(window)

dip_rate = (
    tail["TVT"].diff()
    /
    tail["MD"].diff()
)

print("known:",known["MD"].iloc[-1] - tail["MD"].iloc[0])
print("tail:", tail["MD"].iloc[-1] - tail["MD"].iloc[0])
print("dip_rate:",
    dip_rate
    .replace([np.inf, -np.inf], np.nan)
    .dropna()
    .describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)
```

```python
# This is a very important result and much more relevant than the earlier whole-known-zone dip analysis.

# Let's interpret it carefully.

# What you measured

# For one well, you computed the dip rate over the last ~500 ft before PS:

# Mean   = 0.0087 ft/ft
# Median = 0.0100 ft/ft
# P95    = 0.0200 ft/ft

# Compare that to the whole-known-zone result:

# Mean ≈ 0.42 ft/ft

# The difference is enormous.

# Geological Meaning

# Earlier we suspected:

# Build section
# → Landing
# → Long lateral

# This result confirms it.

# During the final 500 ft before PS:

# 100 ft drilled
# ≈
# 1 ft TVT change

# because:

# 0.01 ft/ft

# is essentially flat.

# Competition Implication

# This is probably one of the strongest findings from all the EDA you've run.

# Near the prediction boundary:

# TVT becomes extremely stable.

# The model is not forecasting large geometric changes.

# It is forecasting:

# small geological drift
# around an almost constant TVT.
# Why this matters

# Suppose:

# Prediction length = 1000 ft

# If the local dip remains:

# 0.01 ft/ft

# then expected drift is only:

# 1000 × 0.01
# =
# 10 ft TVT

# which is tiny relative to:

# TVT ≈ 11000–13000 ft

# The EDA We Actually Need Now
# The single-well result is interesting, but we need the distribution across all wells.
```

```python
# Yes. To properly answer the EDA question, you need to compute the tail dip rate for every well, then look at the distribution.

# This code:

# Uses only the known zone (TVT_input not null)
# Takes the last 500 ft before PS
# Computes dip rates (ΔTVT / ΔMD)
# Stores per-well statistics
# Produces the global distribution

TAIL_WINDOW = 500

results = []
all_tail_dips = []

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for fp in tqdm(horizontal_files):

    well_id = (
        os.path.basename(fp)
        .replace("__horizontal_well.csv", "")
    )

    df = pd.read_csv(fp)

    known = df.loc[
        df["TVT_input"].notna(),
        ["MD", "TVT"]
    ].copy()

    if len(known) < 50:
        continue

    tail = known.tail(
        min(TAIL_WINDOW, len(known))
    )

    dip_rate = (
        tail["TVT"].diff()
        /
        tail["MD"].diff()
    )

    dip_rate = (
        dip_rate
        .replace([np.inf, -np.inf], np.nan)
        .dropna()
    )

    if len(dip_rate) == 0:
        continue

    all_tail_dips.extend(
        dip_rate.values
    )

    results.append({
        "well_id": well_id,

        "known_len": len(known),

        "tail_mean_dip":
            dip_rate.mean(),

        "tail_median_dip":
            dip_rate.median(),

        "tail_std_dip":
            dip_rate.std(),

        "tail_p95_abs_dip":
            np.percentile(
                np.abs(dip_rate),
                95
            ),

        "tail_max_abs_dip":
            np.max(
                np.abs(dip_rate)
            )
    })

tail_df = pd.DataFrame(results)

print("=" * 80)
print("PER-WELL TAIL DIP RATE")
print("=" * 80)

print(
    tail_df["tail_mean_dip"]
    .describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\n")

print("=" * 80)
print("GLOBAL LAST-500FT DIP DISTRIBUTION")
print("=" * 80)

all_tail_dips = pd.Series(all_tail_dips)

print(
    all_tail_dips.describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\nTop 20 steepest wells:")

print(
    tail_df.sort_values(
        "tail_mean_dip",
        ascending=False
    ).head(20)
)

tail_df.to_csv(
    "tail_dip_rate_stats.csv",
    index=False
)
```

# Check TVT continuity at PS boundary
### Does TVT change slope abruptly at PS? Discontinuity = PS is not arbitrary — it marks a geological event.

```python
WINDOW = 100

results = []

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for fp in tqdm(horizontal_files):

    well_id = (
        os.path.basename(fp)
        .replace("__horizontal_well.csv", "")
    )

    df = pd.read_csv(fp)

    ps_idx = df["TVT_input"].last_valid_index()

    if ps_idx is None:
        continue

    # Need enough rows on both sides
    if ps_idx < WINDOW:
        continue

    if ps_idx + WINDOW >= len(df):
        continue

    before = df.iloc[
        ps_idx - WINDOW : ps_idx + 1
    ]

    after = df.iloc[
        ps_idx : ps_idx + WINDOW + 1
    ]

    before_slope = (
        before["TVT"].diff()
        /
        before["MD"].diff()
    ).dropna()

    after_slope = (
        after["TVT"].diff()
        /
        after["MD"].diff()
    ).dropna()

    mean_before = before_slope.mean()
    mean_after = after_slope.mean()

    slope_change = (
        mean_after - mean_before
    )

    abs_slope_change = abs(
        slope_change
    )

    results.append({
        "well_id": well_id,

        "before_slope":
            mean_before,

        "after_slope":
            mean_after,

        "slope_change":
            slope_change,

        "abs_slope_change":
            abs_slope_change
    })

summary = pd.DataFrame(results)

print("=" * 80)
print("PS SLOPE CONTINUITY")
print("=" * 80)

print("\nSlope change distribution:")

print(
    summary["slope_change"]
    .describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\nAbsolute slope change:")

print(
    summary["abs_slope_change"]
    .describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\nTop 20 discontinuities:")

print(
    summary.sort_values(
        "abs_slope_change",
        ascending=False
    )
    .head(20)
)

summary.to_csv(
    "ps_slope_continuity.csv",
    index=False
)
```

# Plot autocorrelation of TVT series.
### How many MD lags before TVT becomes uncorrelated? Determines minimum useful window size.

```python
# Average ACF of dTVT Across Wells

# Computes autocorrelation up to 500 MD lags.

MAX_LAG = 500

acf_sum = np.zeros(MAX_LAG)
acf_count = np.zeros(MAX_LAG)

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for fp in tqdm(horizontal_files):

    df = pd.read_csv(fp)

    known = df.loc[
        df["TVT_input"].notna(),
        "TVT"
    ]

    if len(known) < MAX_LAG + 50:
        continue

    dtvt = known.diff().dropna().values

    if np.std(dtvt) < 1e-8:
        continue

    for lag in range(1, MAX_LAG + 1):

        x = dtvt[:-lag]
        y = dtvt[lag:]

        corr = np.corrcoef(x, y)[0, 1]

        if np.isfinite(corr):
            acf_sum[lag - 1] += corr
            acf_count[lag - 1] += 1

mean_acf = acf_sum / np.maximum(acf_count, 1)

acf_df = pd.DataFrame({
    "lag_ft": np.arange(1, MAX_LAG + 1),
    "mean_acf": mean_acf
})

print(acf_df.head())

plt.figure(figsize=(12, 6))
plt.plot(
    acf_df["lag_ft"],
    acf_df["mean_acf"]
)

plt.axhline(
    0.0,
    linestyle="--"
)

plt.axhline(
    0.2,
    linestyle=":"
)

plt.title(
    "Average Autocorrelation of dTVT"
)

plt.xlabel("Lag (MD ft)")
plt.ylabel("Correlation")

plt.grid(True)
plt.show()

acf_df.to_csv(
    "dtvt_autocorrelation.csv",
    index=False
)
```

# Compute TVT vs Z residuals after linear fit (per well)
### r=-0.999 globally but per-well residual is what LGB must learn. Plot residual distribution.

```python
# Per-Well Linear Fit Residual Analysis

from sklearn.linear_model import LinearRegression

all_residuals = []
well_stats = []

horizontal_files = glob.glob(
    os.path.join(TRAIN_DIR, "*__horizontal_well.csv")
)

for fp in tqdm(horizontal_files):

    well_id = (
        os.path.basename(fp)
        .replace("__horizontal_well.csv", "")
    )

    df = pd.read_csv(fp)

    valid = df[
        ["TVT", "Z"]
    ].dropna()

    if len(valid) < 50:
        continue

    X = valid[["Z"]].values
    y = valid["TVT"].values

    model = LinearRegression()
    model.fit(X, y)

    pred = model.predict(X)

    residual = y - pred

    all_residuals.extend(
        residual.tolist()
    )

    well_stats.append({
        "well_id": well_id,

        "slope": model.coef_[0],
        "intercept": model.intercept_,

        "resid_mean":
            residual.mean(),

        "resid_std":
            residual.std(),

        "resid_mae":
            np.mean(
                np.abs(residual)
            ),

        "resid_max_abs":
            np.max(
                np.abs(residual)
            ),

        "r2":
            model.score(X, y)
    })

well_stats = pd.DataFrame(well_stats)

print("=" * 80)
print("PER-WELL TVT~Z FIT")
print("=" * 80)

print("\nR² distribution")
print(
    well_stats["r2"].describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\nResidual MAE distribution")
print(
    well_stats["resid_mae"].describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

print("\nResidual max abs distribution")
print(
    well_stats["resid_max_abs"].describe(
        percentiles=[
            0.01,
            0.05,
            0.25,
            0.50,
            0.75,
            0.95,
            0.99
        ]
    )
)

well_stats.to_csv(
    "tvt_z_residual_stats.csv",
    index=False
)
```

# 3. Z vs TVT Relationship (Highest Priority)
This section analyzes the relationship between the trajectory height (Z) and geology thickness (TVT).
It answers these crucial questions:
1. Does TVT relate linearly to Z within the known zone? Fits `TVT = a*Z + b` and computes $R^2$.
2. How is the slope `a` distributed across all wells? Is it stable within a single well?
3. What is the RMSE of pure Z-fit extrapolation on the evaluation zone?
4. Can we predict the slope `a` from neighbor wells?
5. Does the residual grow with Measured Depth from Prediction Start?
6. Does the TVT residual after Z-fit correlate with GR? (Validating GR as a correction feature)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from scipy.spatial import cKDTree
from sklearn.metrics import r2_score, mean_squared_error

# Ensure datasets are loaded
if 'horizontal_master' not in globals():
    import glob
    TRAIN_DIR = "../data/train"
    horizontal_files = glob.glob(f"{TRAIN_DIR}/*__horizontal_well.csv")
    well_ids = sorted([os.path.basename(f).split("__")[0] for f in horizontal_files])
    # Fast load subset of 150 wells if not loaded
    print("Loading dataset subset for independent execution...")
    all_horizontal = []
    for wid in tqdm(well_ids[:150]):
        df = pd.read_csv(f"{TRAIN_DIR}/{wid}__horizontal_well.csv")
        df['well_id'] = wid
        all_horizontal.append(df)
    horizontal_master = pd.concat(all_horizontal, ignore_index=True)

# Process per-well
if 'well_id' in horizontal_master.columns:
    well_groups = horizontal_master.groupby('well_id')
else:
    well_groups = horizontal_master.groupby(horizontal_master['id'].apply(lambda x: x.rsplit('_', 1)[0]))

slopes = []
r2_scores = []
slope_stabilities = []
rmse_z_extrap = []
rmse_carry_forward = []
drift_resids = []
gr_corrs = []
spatial_coords = []

for wid, df in tqdm(well_groups):
    df = df.sort_values('MD')
    known = df[df['TVT_input'].notna()]
    pred = df[df['TVT_input'].isna()]
    
    if len(known) < 10 or len(pred) < 10:
        continue
    
    # Fit OLS: TVT = a*Z + b on known zone
    Z_k = known['Z'].values
    TVT_k = known['TVT_input'].values
    a, b = np.polyfit(Z_k, TVT_k, 1)
    
    # R2 on known zone
    pred_k = a * Z_k + b
    r2 = r2_score(TVT_k, pred_k)
    
    # Slope stability (first half vs second half of known zone)
    mid = len(known) // 2
    if mid > 5:
        a1, _ = np.polyfit(Z_k[:mid], TVT_k[:mid], 1)
        a2, _ = np.polyfit(Z_k[mid:], TVT_k[mid:], 1)
        slope_stability = abs(a1 - a2)
    else:
        slope_stability = np.nan
    
    # RMSE on prediction zone
    Z_p = pred['Z'].values
    TVT_p = pred['TVT'].values
    pred_p_z = a * Z_p + b
    rmse_z = np.sqrt(mean_squared_error(TVT_p, pred_p_z))
    
    # Carry-forward baseline
    last_known_tvt = TVT_k[-1]
    rmse_cf = np.sqrt(mean_squared_error(TVT_p, np.full_like(TVT_p, last_known_tvt)))
    
    # Correlation of Z-fit residual with GR (Known Zone)
    gr_k = known['GR'].interpolate('linear', limit_direction='both').ffill().bfill().fillna(90.0).values
    residuals_k = TVT_k - pred_k
    gr_corr = pd.Series(residuals_k).corr(pd.Series(gr_k))
    
    # Drift residual (Z residual vs MD_from_PS)
    md_from_ps = pred['MD'].values - known['MD'].values[-1]
    residuals_p = TVT_p - pred_p_z
    for m, r in zip(md_from_ps, residuals_p):
        drift_resids.append({'md_from_ps': m, 'residual': abs(r)})
        
    slopes.append(a)
    r2_scores.append(r2)
    slope_stabilities.append(slope_stability)
    rmse_z_extrap.append(rmse_z)
    rmse_carry_forward.append(rmse_cf)
    gr_corrs.append(gr_corr)
    spatial_coords.append({'well_id': wid, 'cx': df['X'].mean(), 'cy': df['Y'].mean(), 'slope': a})

df_slopes = pd.DataFrame(spatial_coords)
df_drift = pd.DataFrame(drift_resids)

print("=== QUANTITATIVE ANSWERS (SECTION 3) ===")
print(f"1. Mean R² of Z-fit on known zone: {np.nanmean(r2_scores):.4f}")
print(f"2. Slope (a) across all wells - Mean: {np.nanmean(slopes):.4f}, Std: {np.nanstd(slopes):.4f}")
print(f"3. Median Slope Instability (abs diff first/second half): {np.nanmedian(slope_stabilities):.4f}")
print(f"4. Mean RMSE on eval zone (Pure Z-fit extrapolation): {np.nanmean(rmse_z_extrap):.4f} ft")
print(f"5. Mean RMSE on eval zone (Carry-forward baseline): {np.nanmean(rmse_carry_forward):.4f} ft")
print(f"6. Mean Correlation of Z-fit Residuals with GR (known zone): {np.nanmean(gr_corrs):.4f}")

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.histplot(slopes, ax=axes[0, 0], kde=True, color='royalblue')
axes[0, 0].set_title("Distribution of Slope (a) across Wells")
axes[0, 0].set_xlabel("Slope (a)")

sns.histplot(slope_stabilities, ax=axes[0, 1], kde=True, color='forestgreen')
axes[0, 1].set_title("Slope Instability (Abs Diff 1st vs 2nd Half)")
axes[0, 1].set_xlabel("Delta Slope")

sns.histplot(gr_corrs, ax=axes[1, 0], kde=True, color='purple')
axes[1, 0].set_title("Correlation of Z-fit Residual with GR (Validates GR correction)")
axes[1, 0].set_xlabel("Pearson r")

# Plot drift vs MD_from_PS
df_drift['md_bin'] = (df_drift['md_from_ps'] // 200) * 200
drift_grouped = df_drift.groupby('md_bin')['residual'].mean().reset_index()
axes[1, 1].plot(drift_grouped['md_bin'], drift_grouped['residual'], marker='o', color='crimson')
axes[1, 1].set_title("Mean Absolute Residual Growth vs MD from PS (Drift)")
axes[1, 1].set_xlabel("MD from PS (ft)")
axes[1, 1].set_ylabel("Mean Absolute Residual (ft)")
plt.tight_layout()
plt.show()

# 3.4 Predict slope from neighbor wells
centroids = df_slopes[['cx', 'cy']].values
tree = cKDTree(centroids)
neigh_slopes_pred = []
actual_slopes = df_slopes['slope'].values

for i, (cx, cy) in enumerate(centroids):
    dist, idxs = tree.query([cx, cy], k=4)
    other_idxs = [idx for idx in idxs if idx != i]
    if len(other_idxs) == 0:
        neigh_slopes_pred.append(actual_slopes[i])
    else:
        neigh_slopes_pred.append(np.mean(actual_slopes[other_idxs]))

neigh_corr = pd.Series(actual_slopes).corr(pd.Series(neigh_slopes_pred))
print(f"7. Correlation between target slope and neighbor-mean slope: {neigh_corr:.4f}")
```

```python
from scipy.interpolate import interp1d

sample_wids = list(well_groups.groups.keys())[:5]

# 4.1 Plot GR vs MD for 5 wells
fig, axes = plt.subplots(5, 1, figsize=(14, 10), sharex=True)
for i, wid in enumerate(sample_wids):
    df = well_groups.get_group(wid).sort_values('MD')
    axes[i].plot(df['MD'], df['GR'], color='black', alpha=0.7)
    axes[i].axvline(df[df['TVT_input'].notna()]['MD'].max(), color='red', linestyle='--', label='PS')
    axes[i].set_title(f"Well {wid} - GR vs MD")
    axes[i].set_ylabel("GR (API)")
axes[4].set_xlabel("Measured Depth (MD)")
plt.tight_layout()
plt.show()

# 4.2 GR Distribution Shift
fig, axes = plt.subplots(1, 5, figsize=(16, 3))
for i, wid in enumerate(sample_wids):
    df = well_groups.get_group(wid).sort_values('MD')
    known = df[df['TVT_input'].notna()]
    pred = df[df['TVT_input'].isna()]
    sns.kdeplot(known['GR'].dropna(), ax=axes[i], label='Known', fill=True, color='blue')
    sns.kdeplot(pred['GR'].dropna(), ax=axes[i], label='Eval', fill=True, color='orange')
    axes[i].set_title(f"Well {wid}")
    axes[i].legend()
plt.suptitle("GR Distribution Shift: Known vs Eval Zone")
plt.tight_layout()
plt.show()

# 4.3 GR Autocorrelation
acf_lags_all = []
for wid, df in tqdm(well_groups):
    gr_clean = df['GR'].interpolate('linear', limit_direction='both').ffill().bfill().fillna(90.0).values
    if len(gr_clean) < 100: continue
    gr_dm = gr_clean - np.mean(gr_clean)
    acf = np.correlate(gr_dm, gr_dm, mode='full')
    acf = acf[len(acf)//2:]
    acf /= acf[0]
    lag_20 = np.where(acf < 0.2)[0]
    if len(lag_20) > 0:
        acf_lags_all.append(lag_20[0])
print(f"Median GR autocorrelation lag (where r < 0.2): {np.median(acf_lags_all):.1f} ft")

# 4.4 Slide 9 Correlation check (Own Pre-PS vs Typewell)
pre_ps_corrs = []
typewell_corrs = []
for wid in tqdm(sample_wids):
    try:
        hw = pd.read_csv(f"../data/train/{wid}__horizontal_well.csv").sort_values('MD')
        tw = pd.read_csv(f"../data/train/{wid}__typewell.csv").sort_values('TVT')
    except:
        continue
    ps_idx = int(hw['TVT_input'].notna().values.nonzero()[0][-1])
    known = hw.iloc[:ps_idx+1].dropna(subset=['TVT', 'GR'])
    pred = hw.iloc[ps_idx+1:].dropna(subset=['TVT', 'GR'])
    tw_clean = tw.dropna(subset=['TVT', 'GR'])
    
    # Align Pre-PS and Post-PS in TVT
    min_tvt = max(known['TVT'].min(), pred['TVT'].min())
    max_tvt = min(known['TVT'].max(), pred['TVT'].max())
    if min_tvt < max_tvt:
        tvt_axis = np.arange(min_tvt, max_tvt, 1.0)
        gr_pre = interp1d(known['TVT'], known['GR'], bounds_error=False)(tvt_axis)
        gr_post = interp1d(pred['TVT'], pred['GR'], bounds_error=False)(tvt_axis)
        mask = ~np.isnan(gr_pre) & ~np.isnan(gr_post)
        if mask.sum() > 5:
            pre_ps_corrs.append(pd.Series(gr_pre[mask]).corr(pd.Series(gr_post[mask])))
            
    # Align Post-PS and Typewell in TVT
    min_tvt_tw = max(tw_clean['TVT'].min(), pred['TVT'].min())
    max_tvt_tw = min(tw_clean['TVT'].max(), pred['TVT'].max())
    if min_tvt_tw < max_tvt_tw:
        tvt_axis_tw = np.arange(min_tvt_tw, max_tvt_tw, 1.0)
        gr_tw = interp1d(tw_clean['TVT'], tw_clean['GR'], bounds_error=False)(tvt_axis_tw)
        gr_post_tw = interp1d(pred['TVT'], pred['GR'], bounds_error=False)(tvt_axis_tw)
        mask = ~np.isnan(gr_tw) & ~np.isnan(gr_post_tw)
        if mask.sum() > 5:
            typewell_corrs.append(pd.Series(gr_tw[mask]).corr(pd.Series(gr_post_tw[mask])))

print(f"Mean post-PS GR vs own pre-PS GR correlation in TVT space (Slide 9): {np.nanmean(pre_ps_corrs):.4f}")
print(f"Mean post-PS GR vs typewell GR correlation in TVT space: {np.nanmean(typewell_corrs):.4f}")

# 4.5 GR null rate by MD_from_PS
drift_gr_nans = []
for wid, df in tqdm(well_groups):
    df = df.sort_values('MD')
    known_len = df['TVT_input'].notna().sum()
    pred = df.iloc[known_len:]
    if len(pred) == 0: continue
    md_from_ps = pred['MD'].values - df['MD'].values[known_len-1]
    gr_isnan = pred['GR'].isna().astype(int).values
    for m, nan_val in zip(md_from_ps, gr_isnan):
        drift_gr_nans.append({'md_from_ps': m, 'isnan': nan_val})
df_gr_nan = pd.DataFrame(drift_gr_nans)
df_gr_nan['md_bin'] = (df_gr_nan['md_from_ps'] // 200) * 200
nan_drift = df_gr_nan.groupby('md_bin')['isnan'].mean().reset_index()

plt.figure(figsize=(8, 4))
plt.plot(nan_drift['md_bin'], nan_drift['isnan'], marker='o', color='purple')
plt.title("GR Null Rate vs MD from PS (Drift)")
plt.xlabel("MD from PS (ft)")
plt.ylabel("Missing GR Ratio")
plt.show()

# 4.6 Window correlation score
window_sizes = [10, 30, 50, 100]
window_corrs = {w: [] for w in window_sizes}
for wid, df in tqdm(well_groups):
    df = df.sort_values('MD')
    known_len = df['TVT_input'].notna().sum()
    if known_len < 120 or len(df) - known_len < 120: continue
    gr = df['GR'].interpolate('linear', limit_direction='both').ffill().bfill().fillna(90.0).values
    known_gr = gr[:known_len]
    pred_gr = gr[known_len:]
    for w in window_sizes:
        w_known = known_gr[-w:]
        w_pred = pred_gr[:w]
        r = pd.Series(w_known).corr(pd.Series(w_pred))
        if not np.isnan(r):
            window_corrs[w].append(r)
mean_window_corrs = [np.mean(window_corrs[w]) for w in window_sizes]
plt.figure(figsize=(8, 4))
plt.plot(window_sizes, mean_window_corrs, marker='s', color='orange')
plt.title("Mean GR Window Correlation vs Window Size at PS")
plt.xlabel("Window Size (ft)")
plt.ylabel("Pearson r")
plt.show()
```

# 5. Trajectory & Spatial Features
This section evaluates structural features along the well paths, direction changes, and spatial relationship between wells.
It includes:
1. Confirming MD vs. Y relationship per-well.
2. Distribution of trajectory inclination proxy `dZ/dMD`.
3. Azimuth changes (dog-leg behavior) within the prediction zone.
4. Correlation of Dog-Leg Severity (DLS) vs. TVT residuals.
5. Mapping well centroids and spatial clustering.
6. Confirming if neighboring wells carry predictive dip slopes.


```python
from sklearn.cluster import KMeans

md_y_corrs = []
inclinations = []
azimuth_drifts = []
dls_resid_corrs = []

for wid, df in tqdm(well_groups):
    df = df.sort_values('MD')
    known = df[df['TVT_input'].notna()]
    pred = df[df['TVT_input'].isna()]
    if len(df) < 50: continue
    
    # MD vs Y corr
    r_md_y = df['MD'].corr(df['Y'])
    md_y_corrs.append(r_md_y)
    
    # dZ/dMD (inclination proxy)
    dZ = df['Z'].diff().values
    dMD = df['MD'].diff().values
    incl = dZ / np.maximum(dMD, 0.1)
    inclinations.extend(incl[~np.isnan(incl)])
    
    # DLS vs residuals
    if len(known) > 10:
        a, b = np.polyfit(known['Z'], known['TVT_input'], 1)
        resids = df['TVT'] - (a * df['Z'] + b)
        d_incl = np.diff(incl, prepend=0.0)
        dls = np.abs(d_incl)
        r_dls = pd.Series(dls).corr(pd.Series(resids))
        if not np.isnan(r_dls):
            dls_resid_corrs.append(r_dls)
            
    # Azimuth drift
    if len(pred) > 10:
        dx_k = known['X'].diff().mean()
        dy_k = known['Y'].diff().mean()
        az_k = np.arctan2(dy_k, dx_k)
        dx_p = pred['X'].diff().mean()
        dy_p = pred['Y'].diff().mean()
        az_p = np.arctan2(dy_p, dx_p)
        az_drift = np.abs(az_k - az_p)
        azimuth_drifts.append(az_drift)

print(f"Median MD vs Y correlation across wells: {np.nanmedian(md_y_corrs):.4f}")
print(f"Mean correlation of Dog-Leg Severity (DLS) vs. TVT residuals: {np.nanmean(dls_resid_corrs):.4f}")
print(f"Median Azimuth change between Known and Pred (rad): {np.nanmedian(azimuth_drifts):.4f}")

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(inclinations, ax=axes[0], bins=100, color='teal', kde=True)
axes[0].set_title("Inclination Proxy (dZ/dMD) Distribution")
axes[0].set_xlabel("dZ/dMD")

sns.histplot(azimuth_drifts, ax=axes[1], bins=50, color='darkred', kde=True)
axes[1].set_title("Azimuth Drift (rad)")
axes[1].set_xlabel("Azimuth Delta")
plt.tight_layout()
plt.show()

# Well centroid map & clustering
plt.figure(figsize=(10, 8))
kmeans = KMeans(n_clusters=5, random_state=42, n_init='auto')
df_slopes['cluster'] = kmeans.fit_predict(df_slopes[['cx', 'cy']])
sns.scatterplot(data=df_slopes, x='cx', y='cy', hue='cluster', palette='Set1', s=40)
plt.title("Spatial Centroids Map & K-Means Clustering")
plt.xlabel("X Coordinate (ft)")
plt.ylabel("Y Coordinate (ft)")
plt.show()
```

# 8. Feature Engineering Validation
In this section, we validate the mathematical logic of the LightGBM target and coordinate exclusions:
1. Confirming if `TVT_z_fit` residuals represent a stable, smooth target for LightGBM.
2. Testing if rolling GR mean (31ft) correlates with the Z-fit residual.
3. Validating if `md_from_ps` correlates with absolute residual magnitude (validates adding $MD^2$).
4. Comparing rolling Z-fit vs. global per-well Z-fit extrapolation RMSE.


```python
resids_all = []
gr_roll_mean_corrs = []
md_resid_corrs = []
rmse_rolling_z = []
rmse_global_z = []

for wid, df in tqdm(well_groups):
    df = df.sort_values('MD')
    known = df[df['TVT_input'].notna()]
    pred = df[df['TVT_input'].isna()]
    if len(known) < 50 or len(pred) < 50: continue
    
    # Global Z fit
    a_k, b_k = np.polyfit(known['Z'], known['TVT_input'], 1)
    pred_resids = pred['TVT'] - (a_k * pred['Z'] + b_k)
    resids_all.extend(pred_resids)
    
    # 8.1 Rolling GR vs residuals
    gr = pred['GR'].interpolate('linear', limit_direction='both').ffill().bfill().fillna(90.0)
    gr_roll = gr.rolling(31, center=True, min_periods=1).mean()
    r_gr_resid = gr_roll.corr(pred_resids)
    if not np.isnan(r_gr_resid):
        gr_roll_mean_corrs.append(r_gr_resid)
        
    # 8.2 MD from PS vs absolute residual
    md_from_ps = pred['MD'] - known['MD'].iloc[-1]
    r_md_res = md_from_ps.corr(np.abs(pred_resids))
    if not np.isnan(r_md_res):
        md_resid_corrs.append(r_md_res)
        
    # 8.3 Rolling Z fit RMSE (sliding 200ft window fit)
    pred_z_rolling = []
    for md_val in pred['MD']:
        window = known[known['MD'] >= md_val - 200]
        if len(window) < 10: window = known
        a_w, b_w = np.polyfit(window['Z'], window['TVT_input'], 1)
        pred_z_rolling.append(a_w * df[df['MD'] == md_val]['Z'].iloc[0] + b_w)
        
    rmse_roll = np.sqrt(mean_squared_error(pred['TVT'], pred_z_rolling))
    rmse_glob = np.sqrt(mean_squared_error(pred['TVT'], a_k * pred['Z'] + b_k))
    rmse_rolling_z.append(rmse_roll)
    rmse_global_z.append(rmse_glob)

print("=== FEATURE ENGINEERING METRICS ===")
print(f"Mean correlation between rolling GR mean (31ft) and Z-fit residual: {np.nanmean(gr_roll_mean_corrs):.4f}")
print(f"Mean correlation between MD from PS and absolute residual (validates MD²): {np.nanmean(md_resid_corrs):.4f}")
print(f"Mean Rolling 200ft Z-fit RMSE: {np.mean(rmse_rolling_z):.4f} ft")
print(f"Mean Global per-well Z-fit RMSE: {np.mean(rmse_global_z):.4f} ft")

# Plot residual distribution
plt.figure(figsize=(8, 4))
sns.histplot(resids_all, bins=100, color='darkorange', kde=True)
plt.title("LightGBM Target: TVT Z-Fit Residuals (Smooth & Centered)")
plt.xlabel("Residual (ft)")
plt.show()
```

### Check A: GR Log Calibration & Normalization (Petrophysical Scaling)
We evaluate if per-wellbore log standardization (Z-score) or Min-Max scaling alters or improves the Pearson correlation between horizontal known GR and vertical typewell GR.
*(Note: Pearson's correlation coefficient is mathematically invariant under linear transformations, but aligning scale offsets remains crucial for direct difference features in ML).* 


```python
# Check A: GR Calibration & Normalization Comparison
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
from scipy.interpolate import interp1d

try:
    TRAIN_DIR = "../data/train"
    wids = sorted(set(f.split('__')[0] for f in os.listdir(TRAIN_DIR) if '__horizontal_well.csv' in f))
except:
    TRAIN_DIR = "data/train"
    wids = sorted(set(f.split('__')[0] for f in os.listdir(TRAIN_DIR) if '__horizontal_well.csv' in f))

subset_wids = wids[:150]  # Fast representative subset
corr_raw = []
corr_std = []
corr_minmax = []

def robust_scale(series):
    q10, q90 = series.quantile(0.1), series.quantile(0.9)
    if q90 == q10: return series - q10
    return (series - q10) / (q90 - q10)
    
for wid in tqdm(subset_wids):
    try:
        hw = pd.read_csv(f"{TRAIN_DIR}/{wid}__horizontal_well.csv").sort_values('MD')
        tw = pd.read_csv(f"{TRAIN_DIR}/{wid}__typewell.csv").sort_values('TVT')
    except:
        continue
        
    known = hw[hw['TVT_input'].notna()].dropna(subset=['TVT_input', 'GR'])
    tw_clean = tw.dropna(subset=['TVT', 'GR'])
    
    if len(known) > 10 and len(tw_clean) > 10:
        min_tvt = max(known['TVT_input'].min(), tw_clean['TVT'].min())
        max_tvt = min(known['TVT_input'].max(), tw_clean['TVT'].max())
        if min_tvt < max_tvt:
            tvt_axis = np.arange(min_tvt, max_tvt, 1.0)
            gr_hw_raw = interp1d(known['TVT_input'], known['GR'], bounds_error=False)(tvt_axis)
            gr_tw_raw = interp1d(tw_clean['TVT'], tw_clean['GR'], bounds_error=False)(tvt_axis)
            mask = ~np.isnan(gr_hw_raw) & ~np.isnan(gr_tw_raw)
            if mask.sum() > 5:
                corr_raw.append(pd.Series(gr_hw_raw[mask]).corr(pd.Series(gr_tw_raw[mask])))
                
                gr_hw_std = (known['GR'] - known['GR'].mean()) / max(known['GR'].std(), 1e-6)
                gr_tw_std = (tw_clean['GR'] - tw_clean['GR'].mean()) / max(tw_clean['GR'].std(), 1e-6)
                gr_hw_std_interp = interp1d(known['TVT_input'], gr_hw_std, bounds_error=False)(tvt_axis)
                gr_tw_std_interp = interp1d(tw_clean['TVT'], gr_tw_std, bounds_error=False)(tvt_axis)
                corr_std.append(pd.Series(gr_hw_std_interp[mask]).corr(pd.Series(gr_tw_std_interp[mask])))
                
                gr_hw_mm = robust_scale(known['GR'])
                gr_tw_mm = robust_scale(tw_clean['GR'])
                gr_hw_mm_interp = interp1d(known['TVT_input'], gr_hw_mm, bounds_error=False)(tvt_axis)
                gr_tw_mm_interp = interp1d(tw_clean['TVT'], gr_tw_mm, bounds_error=False)(tvt_axis)
                corr_minmax.append(pd.Series(gr_hw_mm_interp[mask]).corr(pd.Series(gr_tw_mm_interp[mask])))
                
print(f"Mean Correlation - Raw API GR: {np.nanmean(corr_raw):.4f}")
print(f"Mean Correlation - Z-Score Standardized: {np.nanmean(corr_std):.4f}")
print(f"Mean Correlation - Robust Min-Max Scaled: {np.nanmean(corr_minmax):.4f}")
```

### Check B: Type Log Out-of-Bounds (OOB) Extrapolation Risk
We simulate the baseline dipping plane (`tvt_plane`) and check if the horizontal prediction zone's trajectory climbs or drops outside the vertical typewell logged TVT range, causing edge flatlines.


```python
# Check B: Type Log Out-of-Bounds (OOB) Extrapolation Risk
oob_percentages = []

for wid in tqdm(subset_wids):
    try:
        hw = pd.read_csv(f"{TRAIN_DIR}/{wid}__horizontal_well.csv").sort_values('MD')
        tw = pd.read_csv(f"{TRAIN_DIR}/{wid}__typewell.csv").sort_values('TVT')
    except:
        continue
        
    dmd = hw['MD'].diff().fillna(1).replace(0, 1).abs().values
    dx = hw['X'].diff().fillna(0).values
    dy = hw['Y'].diff().fillna(0).values
    incl_proxy = np.sqrt(dx*dx + dy*dy) / np.maximum(dmd, 1e-6)
    lat_mask = incl_proxy > 0.8
    
    hw_lat = hw[lat_mask] if lat_mask.sum() >= 20 else hw
    hw_known_lat = hw_lat[hw_lat['TVT_input'].notna()]
    
    if len(hw_known_lat) >= 10:
        X_loc = hw_known_lat[['X', 'Y']].values
        y_loc = (hw_known_lat['Z'] + hw_known_lat['TVT_input']).values
        Xb = np.hstack([X_loc, np.ones((X_loc.shape[0], 1))])
        beta = np.linalg.inv(Xb.T @ Xb + 0.1 * np.eye(3)) @ Xb.T @ y_loc
        
        last_row = hw[hw['TVT_input'].notna()].iloc[-1]
        C = (last_row['Z'] + last_row['TVT_input']) - (beta[0] * last_row['X'] + beta[1] * last_row['Y'])
        
        tvt_plane = (beta[0] * hw['X'].values + beta[1] * hw['Y'].values + C) - hw['Z'].values
        pred_indices = hw[hw['TVT_input'].isna()].index
        
        if len(pred_indices) > 0:
            pred_tvt_plane = tvt_plane[pred_indices]
            tw_tvt = tw['TVT'].dropna()
            oob_mask = (pred_tvt_plane < tw_tvt.min()) | (pred_tvt_plane > tw_tvt.max())
            oob_percentages.append(oob_mask.mean())
            
oob_percentages = np.array(oob_percentages)
print(f"Wells with ANY prediction rows Out-of-Bounds: {(oob_percentages > 0).mean()*100:.1f}%")
print(f"Mean Prediction Zone Rows Out-of-Bounds: {np.mean(oob_percentages)*100:.1f}%")
```

### Check C: Continuous Missing GR Run Length Distribution
We calculate the continuous run lengths of missing GR data in the prediction zones to determine if simple linear interpolation is a robust missing data strategy.


```python
# Check C: Continuous Missing GR Run Lengths
nan_run_lengths = []

for wid in tqdm(subset_wids):
    try:
        hw = pd.read_csv(f"{TRAIN_DIR}/{wid}__horizontal_well.csv").sort_values('MD')
    except:
        continue
        
    pred_gr = hw[hw['TVT_input'].isna()]['GR']
    if len(pred_gr) > 0:
        nan_mask = pred_gr.isna().values
        run = 0
        for val in nan_mask:
            if val:
                run += 1
            else:
                if run > 0:
                    nan_run_lengths.append(run)
                    run = 0
        if run > 0:
            nan_run_lengths.append(run)
            
nan_run_lengths = np.array(nan_run_lengths)
if len(nan_run_lengths) > 0:
    print(f"Median NaN GR gap length: {np.median(nan_run_lengths):.1f} ft")
    print(f"Mean NaN GR gap length: {np.mean(nan_run_lengths):.1f} ft")
    print(f"99th Percentile NaN GR gap length: {np.percentile(nan_run_lengths, 99):.1f} ft")
    print(f"Max continuous NaN GR gap length: {np.max(nan_run_lengths):.1f} ft")
```
